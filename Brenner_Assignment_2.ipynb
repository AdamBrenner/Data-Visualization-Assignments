```python
# Adam Brenner: Assignment 2

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# PART 1

atomic_names = ['Hydrogen', 'Helium', 'Lithium', 'Beryllium', 'Boron', 'Carbon']

atomic_weights = [1, 4, 7, 9, 11, 12]

explode1 = (0, 0, .2, 0, 0, 0)
explode2 = (0, 0, 0, 0, .2, 0)

fig = plt.figure(figsize = (10,10))

ax1 = fig.add_subplot(121)
ax1.pie(atomic_weights, explode = explode1, labels = atomic_names, autopct='%1.1f%%', shadow = True, startangle = 0)
ax1.axis('equal')
ax1.set_title('Percentages', y = .765)

ax2 = fig.add_subplot(122)
ax2.pie(atomic_weights, explode = explode2, labels = atomic_names, shadow = True,
        autopct=lambda p: '{:.0f}'.format(p * sum(atomic_weights) / 100), startangle = 0)
ax2.axis('equal')
ax2.set_title('Weights', y = .765)

plt.suptitle('Atomic Percentages and Weights of First 6 Elements', fontweight = 'bold', y = .85)
plt.tight_layout(pad = 2)
plt.show()
```

```python
# PART 3

days = ['January 4 2020', 'January 6 2020', 'January 8 2020', 'January 10 2020', 'January 12 2020', 'January 14 2020',
        'January 16 2020', 'January 18 2020']

rand_floats = np.random.uniform(100, 201, 8)

df2 = pd.DataFrame(data = zip(days, rand_floats), columns = ['Days', 'Random'])
df2['Days'] = pd.to_datetime(df2['Days']).dt.date
df2 = df2.set_index(df2['Days'])

fig, axes = plt.subplots(1, 2, figsize = (16,5))
fig.suptitle('Random Values vs. Dates', y = .95)
ax1 = df2.plot(ax = axes[0], color = 'red', kind = 'line')
ax1.set_xticklabels(df2.index, rotation = 45, horizontalalignment = 'right')

ax2 = df2.plot(ax = axes[1], color = 'blue', kind = 'bar')
ax2.set_xticklabels(df2.index, rotation = 45, horizontalalignment = 'right')

ax1
ax2
```

```python
# PART 4

import yfinance as yf

ford = yf.Ticker('F')
# ^ returns a named tuple of Ticker objects

# access each ticker using (example)

ford_df = pd.DataFrame(ford.history(period="1mo"))

fig, axs = plt.subplots(2, 1, sharex=True, figsize = (16, 6))
fig.subplots_adjust(hspace=0)
fig.suptitle('Ford Closing and Volume for 12/28/2020 - 01/27/2021')
ax1 = ford_df.plot(ax = axs[0], y = 'Close', color = 'red', kind = 'line')
ax1.set_ylabel('Close')
ax2 = ford_df.plot(ax = axs[1], y = 'Volume', color = 'black', kind = 'line')
ax2.set_ylabel('Volume')
ax1
ax2
```
